# Доверительный интервалы

## Маршалкин Никита

```r
a <- 1
sd <- 2
alpha <- 0.05
```

```r
ns <- seq(0, 1000, 10)
```

```r
conf.inter <- function(data, a, alpha) {
    left <- length(data) * var(data) / qchisq(1 - alpha / 2, length(data))
    right <- length(data) * var(data) / qchisq(alpha / 2, length(data))
    return (list(left=left, right=right))
}
```

```r
len <- rep(NA, length(ns))
for (i in 1:length(ns)) {
    data <- rnorm(ns[i], a, sd)
    conf.int <- conf.inter(data, a, alpha)
    len[i] <- conf.int$right - conf.int$left
}
```

```r
plot (ns, len)
```
