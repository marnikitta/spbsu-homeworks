```r
theta <- 1
n <- 50
```

## Exponential distribution

```r
ks <- seq(1, 10, 1)
```

```r
sd.k.exp <- rep(NA, length(ks))
```

```r
for (i in 1:length(ks)) {
    k <- ks[i]
    appr.exp <- rep(NA, 1000)
    for (r in 1:1000) {
        data <- rexp(n, rate = theta)
        mom <- mean(data^k)
        appr.exp[r] <- (mom / gamma(k+1)) ^ (1 / k)
    }
    sd.k.exp[i] <- mean((appr.exp - theta)^2)

}
```

```r
plot(ks, sd.k.exp, type = "b")
```

## Unif

```r
ks <- seq(1, 10, 1)
```

```r
sd.k.unif <- rep(NA, length(ks))
```

```r
for (i in 1:length(ks)) {
    k <- ks[i]
    appr.unif <- rep(NA, 1000)
    for (r in 1:1000) {
        data <- runif(n, 0, theta)
        mom <- mean(data^k)
        appr.exp[r] <- (mom * (k + 1)) ^ (1 / k)
    }
    sd.k.unif[i] <- mean((appr.exp - theta)^2)

}
```

```r
plot(ks, sd.k.unif, type = "b")
```
